```python
import tkinter as tk
from tkinter import *
import tkinter.ttk as ttk
import openpyxl
from tkcalendar import Calendar
import pandas as pd

def search():

    sheet_date="2022-05-"+lb_day.get()   #選擇日期
    rank=pd.read_excel("Booking+Agoda.xlsx", sheet_name=sheet_date) #pandas讀取工作表
    df = pd.DataFrame(rank)  #轉成二維pd
    new_df=df.sort_values(["推薦指數"],ascending=0)  #排序
    next_y=0   #為了顯示位置:y軸要+30


    for rank3 in range(3):   #印出推薦前三名飯店的標誌

        # 飯店資訊

        source1=tk.Label(win,text=new_df.at[new_df.index[rank3],"資料來源"],anchor=NW) #資料來源
        source1.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13),width=10)
        source1.place(x=120,y=320+next_y)  #此時y軸會隨著迴圈增加30


        name1=tk.Label(win,text=new_df.at[new_df.index[rank3],"飯店名稱"],anchor=NW) #飯店名稱
        name1.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13),width=15,bd=5)
        name1.place(x=245,y=320+next_y)


        address1=tk.Label(win,text=new_df.at[new_df.index[rank3],"地址"],anchor=NW) #地址
        address1.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13),width=30)
        address1.place(x=440,y=320+next_y)


        check_in1=tk.Label(win,text=new_df.at[new_df.index[rank3],"入住日期"],anchor=NW) #入住日期
        check_in1.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13),width=10)
        check_in1.place(x=790,y=320+next_y)


        roomtype1=tk.Label(win,text=new_df.at[new_df.index[rank3],"房型"],anchor=NW) #房型
        roomtype1.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13),width=15)
        roomtype1.place(x=940,y=320+next_y)


        price1=tk.Label(win,text=new_df.at[new_df.index[rank3],"價格(含早餐)"],anchor=NW)  #價格(含早餐)
        price1.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13),width=5)
        price1.place(x=1110,y=320+next_y)


        recommend1=tk.Label(win,text=new_df.at[new_df.index[rank3],"推薦指數"],anchor=NW) #推薦指數
        recommend1.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13),width=5)
        recommend1.place(x=1225,y=320+next_y)
        next_y+=70



    

# 主視窗
win=tk.Tk()
#視窗屬性設定
win.title('訂房好幫手')
win.geometry('1350x700')
win.iconbitmap('hotel (1).ico')
win.config(bg="#E8E8E8")
win.attributes('-alpha',1)
win.attributes('-topmost',True)
win.maxsize

# 添加背景
canvas = Canvas(width=1350, height=700, highlightthickness=0, borderwidth=0)
canvas.place(x=0, y=0)
bg = PhotoImage(file='11081.png')
bgid = canvas.create_image(0, 0, image=bg, anchor='nw')


# 篩選條件:地區
place = ["選擇你想居住的地區",
         "台北市",
         "新北市",
         "桃園市"]                      
comboExample = ttk.Combobox(win, width=50, font=('微軟正黑體',16))
comboExample["values"] = place
comboExample.place(x=350,y=100)
comboExample.current(0)
# print(dict(comboExample)) 
# print(comboExample.get())


# 篩選條件: 年(固定2022年)
lb_year=tk.Label(text="2022年")
lb_year.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',16))
lb_year.place(x=350,y=150)

# 篩選條件: 月(固定5月)
lb_month=tk.Label(text="5月")
lb_month.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',16))
lb_month.place(x=450,y=150)

# 篩選條件: 日
day_list = ["選擇入住日期","01", "02", "03", "04", "05", "06", "07", "08", "09", "10"]
lb_day = ttk.Combobox(win, width=12, font=('微軟正黑體',16))
lb_day["values"] = day_list
lb_day.place(x=500,y=150)
lb_day.current(0)
# print(dict(lb_day)) 
# print(lb_day.get())

# 篩選條件: 顯示"日"
lb_month=tk.Label(text="日")
lb_month.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',16))
lb_month.place(x=700,y=150)


#按鈕物件:查詢推薦飯店
btn=tk.Button(text='查詢')
img=tk.PhotoImage(file="hotel.png") 
btn.config(image=img,bg="#FFFFFF",fg="#1C1C1C",font=('微軟正黑體',16),width=80,height=80, command=search)
btn.place(x=1000,y=90)

#標籤物件:飯店資訊
lb2=tk.Label(text="推薦:")
lb2.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',16))
lb2.place(x=140,y=222)

#標籤物件:欄位名稱
source=tk.Label(text="資料來源")
source.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13))
source.place(x=120,y=280)

hotel_name=tk.Label(text="飯店名稱")
hotel_name.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13))
hotel_name.place(x=250,y=280)

address=tk.Label(text="地址")
address.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13))
address.place(x=520,y=280)

check_in=tk.Label(text="入住日期")
check_in.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13))
check_in.place(x=800,y=280)

roomtype=tk.Label(text="房型")
roomtype.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13))
roomtype.place(x=960,y=280)

price=tk.Label(text="價格")
price.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13))
price.place(x=1115,y=280)

recommend=tk.Label(text="推薦指數")
recommend.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',13))
recommend.place(x=1200,y=280)


#標籤物件:推薦飯店
lb1=tk.Label(text="以上價格皆含有早餐，最終更新: 2022-04-30")
lb1.config(bg="#FFFFFF",fg="#1C1C1C", font=('微軟正黑體',12))
lb1.place(x=960,y=580)



win.mainloop()
```
